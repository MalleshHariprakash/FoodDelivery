```python
Weather_conditions_Map={"Fog":1,"Sunny":2,"Windy":3,"Sandstorms":4,"Cloudy":5,"Stormy":6}
df['Weather_conditions']=df['Weather_conditions'].replace(Weather_conditions_Map)
```

```python
df['Road_traffic_density'].unique()
```

```python
Road_Traffic_Map={"Low":1,"Medium":2,"High":3,"Jam":4}
df['Road_traffic_density']=df['Road_traffic_density'].replace(Road_Traffic_Map)
```

```python
df['Type_of_order'].unique()
```

```python
Type_of_order_map={"Drinks":1,"Snack":2,"Meal":3,"Buffet":4}
df['Type_of_order']=df['Type_of_order'].replace(Type_of_order_map)
```

```python
df['Type_of_vehicle'].unique()
```

```python
Type_of_vehicle_map={"bicycle":1,"electric_scooter":2,"scooter":3,"motorcycle":4}
df['Type_of_vehicle']=df['Type_of_vehicle'].map(Type_of_vehicle_map)
```

```python
df['Festival'].unique()
```

```python
Festival_Map={"No":0,"Yes":1}
df['Festival']=df['Festival'].map(Festival_Map)
```

```python
df['City'].unique()
```

```python
City_Map={"Semi-Urban":1,"Urban":2,"Metropolitian":3}
df['City']=df['City'].map(City_Map)
```
